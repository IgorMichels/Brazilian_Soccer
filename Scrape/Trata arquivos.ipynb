```python
def catch_teams(text):
    return re.findall('Jogo: (\D+ \/ [A-Z]{2}) X (\D+ \/ [A-Z]{2})', text)
```

```python
files = sorted(glob('*\*\*\*.csv'))
clubs = []
emptys_clubs = []

for i, file in enumerate(files):
    if i % 1000 == 999:
        clear_output()
        print(i + 1)
    
    f = open(file)
    data = f.readlines()
    f.close()
    text = ''
    for row in data:
        text += row

    game_clubs = re.findall('Jogo: (\D+ \/ [A-Z]{2}) X (\D+ \/ [A-Z]{2})', text)
    if len(game_clubs) != 1:
        emptys_clubs.append(file)
    
    clubs += game_clubs
    
clear_output()
len(emptys_clubs)
```

```python
players = []
find_players_manually = []

for i, file in enumerate(files):
    if i % 1000 == 999:
        clear_output()
        print(i + 1)
    
    f = open(file)
    data = f.readlines()
    f.close()
    text = ''
    for row in data:
        text += row

    game_players = re.findall('(\d+\D+[P|A|)|T|R]\s*\d{6})', text)
    if len(game_players) < 22:
        find_players_manually.append(file)
    
    players += game_players
    
clear_output()
len(find_players_manually), len(files)
```

```python
f = open(find_players_manually[0])
data = f.readlines()
f.close()
text = ''
for row in data:
    text += row
    
print(text)
```

```python
def catch_players(text):
    club_1, club_2 = catch_teams(text)[0]
    club = club_1
    players = re.findall('(\d+\D+[P|A|)|T|R]\s*\d{6})', text)
    for i in range(len(players)):
        if i > 0 and 'T(g)' in players[i]:
            club = club_2
        
        players[i] = [players[i], club]
            
    return players
```

```python
f = open(find_players_manually[1])
data = f.readlines()
f.close()
text = ''
for row in data:
    text += row
    
print(catch_players(text))
```
